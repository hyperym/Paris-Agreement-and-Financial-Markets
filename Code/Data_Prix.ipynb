```python


import yfinance as yf
import pandas as pd
from datetime import datetime

# =============================================================================
# LISTE DES ENTREPRISES PAR PAYS
# =============================================================================

companies = {
    # CHINE
    "LONGi Green Energy": {"ticker": "601012.SS", "country": "Chine", "sector": "Green"},
    "BYD Company": {"ticker": "002594.SZ", "country": "Chine", "sector": "Green"},
    "Xinyi Solar": {"ticker": "0968.HK", "country": "Chine", "sector": "Green"},
    "PetroChina": {"ticker": "601857.SS", "country": "Chine", "sector": "Oil_Gas"},
    "CNOOC": {"ticker": "0883.HK", "country": "Chine", "sector": "Oil_Gas"},
    
    # ÉTATS-UNIS
    "NextEra Energy": {"ticker": "NEE", "country": "USA", "sector": "Green"},
    "First Solar": {"ticker": "FSLR", "country": "USA", "sector": "Green"},
    "Ormat Technologies": {"ticker": "ORA", "country": "USA", "sector": "Green"},
    "ExxonMobil": {"ticker": "XOM", "country": "USA", "sector": "Oil_Gas"},
    "Chevron": {"ticker": "CVX", "country": "USA", "sector": "Oil_Gas"},
    
    # RUSSIE
    "RusHydro": {"ticker": "HYDR.ME", "country": "Russie", "sector": "Green"},
    "Inter RAO": {"ticker": "IRAO.ME", "country": "Russie", "sector": "Green"},
    "Gazprom": {"ticker": "GAZP.ME", "country": "Russie", "sector": "Oil_Gas"},
    "Rosneft": {"ticker": "ROSN.ME", "country": "Russie", "sector": "Oil_Gas"},
    "Lukoil": {"ticker": "LKOH.ME", "country": "Russie", "sector": "Oil_Gas"},
    
    # INDE
    "Tata Power": {"ticker": "TATAPOWER.NS", "country": "Inde", "sector": "Green"},
    "NHPC Limited": {"ticker": "NHPC.NS", "country": "Inde", "sector": "Green"},
    "Reliance Industries": {"ticker": "RELIANCE.NS", "country": "Inde", "sector": "Oil_Gas"},
    "ONGC": {"ticker": "ONGC.NS", "country": "Inde", "sector": "Oil_Gas"},
    "Indian Oil Corp": {"ticker": "IOC.NS", "country": "Inde", "sector": "Oil_Gas"},
    
    # JAPON
    "Renova Inc": {"ticker": "9519.T", "country": "Japon", "sector": "Green"},
    "West Holdings": {"ticker": "1407.T", "country": "Japon", "sector": "Green"},
    "ENEOS Holdings": {"ticker": "5020.T", "country": "Japon", "sector": "Oil_Gas"},
    "Inpex Corporation": {"ticker": "1605.T", "country": "Japon", "sector": "Oil_Gas"},
    "Cosmo Energy": {"ticker": "5021.T", "country": "Japon", "sector": "Oil_Gas"},
    
    # ALLEMAGNE
    "Nordex": {"ticker": "NDX1.DE", "country": "Allemagne", "sector": "Green"},
    "SMA Solar": {"ticker": "S92.DE", "country": "Allemagne", "sector": "Green"},
    "Verbio": {"ticker": "VBK.DE", "country": "Allemagne", "sector": "Green"},
    "RWE": {"ticker": "RWE.DE", "country": "Allemagne", "sector": "Oil_Gas"},
    "E.ON": {"ticker": "EOAN.DE", "country": "Allemagne", "sector": "Oil_Gas"},
    
    # BRÉSIL
    "Engie Brasil": {"ticker": "EGIE3.SA", "country": "Bresil", "sector": "Green"},
    "CPFL Energia": {"ticker": "CPFE3.SA", "country": "Bresil", "sector": "Green"},
    "Petrobras": {"ticker": "PETR4.SA", "country": "Bresil", "sector": "Oil_Gas"},
    "PetroRio": {"ticker": "PRIO3.SA", "country": "Bresil", "sector": "Oil_Gas"},
    "Ultrapar": {"ticker": "UGPA3.SA", "country": "Bresil", "sector": "Oil_Gas"},
    
    # CANADA
    "Algonquin Power": {"ticker": "AQN.TO", "country": "Canada", "sector": "Green"},
    "Northland Power": {"ticker": "NPI.TO", "country": "Canada", "sector": "Green"},
    "TransAlta": {"ticker": "TA.TO", "country": "Canada", "sector": "Green"},
    "Suncor Energy": {"ticker": "SU.TO", "country": "Canada", "sector": "Oil_Gas"},
    "Canadian Natural Resources": {"ticker": "CNQ.TO", "country": "Canada", "sector": "Oil_Gas"},
    
    # CORÉE DU SUD
    "Hanwha Solutions": {"ticker": "009830.KS", "country": "Coree_du_Sud", "sector": "Green"},
    "OCI Holdings": {"ticker": "010060.KS", "country": "Coree_du_Sud", "sector": "Green"},
    "SK Innovation": {"ticker": "096770.KS", "country": "Coree_du_Sud", "sector": "Oil_Gas"},
    "S-Oil Corporation": {"ticker": "010950.KS", "country": "Coree_du_Sud", "sector": "Oil_Gas"},
    "GS Holdings": {"ticker": "078930.KS", "country": "Coree_du_Sud", "sector": "Oil_Gas"},
    
    # MEXIQUE
    "Cemex": {"ticker": "CX", "country": "Mexique", "sector": "Green"},
    "Grupo Carso": {"ticker": "GCARSOA1.MX", "country": "Mexique", "sector": "Oil_Gas"},
    "Orbia": {"ticker": "ORBIA.MX", "country": "Mexique", "sector": "Oil_Gas"},
    "Alpek": {"ticker": "ALPEKA.MX", "country": "Mexique", "sector": "Oil_Gas"},
    "Grupo Mexico": {"ticker": "GMEXICOB.MX", "country": "Mexique", "sector": "Oil_Gas"},
    
    # ROYAUME-UNI
    "SSE": {"ticker": "SSE.L", "country": "UK", "sector": "Green"},
    "Drax Group": {"ticker": "DRX.L", "country": "UK", "sector": "Green"},
    "Centrica": {"ticker": "CNA.L", "country": "UK", "sector": "Green"},
    "BP": {"ticker": "BP.L", "country": "UK", "sector": "Oil_Gas"},
    "Shell": {"ticker": "SHEL.L", "country": "UK", "sector": "Oil_Gas"},
    
    # INDONÉSIE
    "Medco Energi": {"ticker": "MEDC.JK", "country": "Indonesie", "sector": "Oil_Gas"},
    "AKR Corporindo": {"ticker": "AKRA.JK", "country": "Indonesie", "sector": "Oil_Gas"},
    "Elnusa": {"ticker": "ELSA.JK", "country": "Indonesie", "sector": "Oil_Gas"},
    "Barito Pacific": {"ticker": "BRPT.JK", "country": "Indonesie", "sector": "Oil_Gas"},
    "Bukit Asam": {"ticker": "PTBA.JK", "country": "Indonesie", "sector": "Oil_Gas"},
    
    # AFRIQUE DU SUD
    "Sasol": {"ticker": "SOL.JO", "country": "Afrique_du_Sud", "sector": "Oil_Gas"},
    "Exxaro Resources": {"ticker": "EXX.JO", "country": "Afrique_du_Sud", "sector": "Oil_Gas"},
    "Impala Platinum": {"ticker": "IMP.JO", "country": "Afrique_du_Sud", "sector": "Oil_Gas"},
    "Kumba Iron Ore": {"ticker": "KIO.JO", "country": "Afrique_du_Sud", "sector": "Oil_Gas"},
    "Anglo American": {"ticker": "AGL.JO", "country": "Afrique_du_Sud", "sector": "Oil_Gas"},
    
    # FRANCE
    "Engie": {"ticker": "ENGI.PA", "country": "France", "sector": "Green"},
    "Voltalia": {"ticker": "VLTSA.PA", "country": "France", "sector": "Green"},
    "Rubis": {"ticker": "RUI.PA", "country": "France", "sector": "Green"},
    "TotalEnergies": {"ticker": "TTE.PA", "country": "France", "sector": "Oil_Gas"},
    "Maurel & Prom": {"ticker": "MAU.PA", "country": "France", "sector": "Oil_Gas"},
    
    # AUSTRALIE
    "AGL Energy": {"ticker": "AGL.AX", "country": "Australie", "sector": "Green"},
    "Origin Energy": {"ticker": "ORG.AX", "country": "Australie", "sector": "Green"},
    "APA Group": {"ticker": "APA.AX", "country": "Australie", "sector": "Green"},
    "Beach Energy": {"ticker": "BPT.AX", "country": "Australie", "sector": "Oil_Gas"},
    "Santos": {"ticker": "STO.AX", "country": "Australie", "sector": "Oil_Gas"},
}

# =============================================================================
# PARAMÈTRES
# =============================================================================

START_DATE = "2013-01-01"
END_DATE = "2018-12-31"

# =============================================================================
# TÉLÉCHARGEMENT DES DONNÉES
# =============================================================================

def download_all_prices():
    """Télécharge les prix de toutes les entreprises"""
    
    all_series = []
    failed = []
    
    print(f"Téléchargement de {len(companies)} actions...")
    print(f"Période : {START_DATE} à {END_DATE}\n")
    
    for name, info in companies.items():
        ticker = info["ticker"]
        try:
            # Télécharger un ticker à la fois
            stock = yf.Ticker(ticker)
            data = stock.history(start=START_DATE, end=END_DATE)
            
            if data.empty or len(data) < 10:
                print(f" {ticker} ({name}): Données insuffisantes")
                failed.append(ticker)
                continue
            
            # Utiliser Close (history() ne retourne pas toujours Adj Close)
            series = data["Close"].copy()
            series.name = ticker
            all_series.append(series)
            print(f"  {ticker}: {len(data)} obs")
                
        except Exception as e:
            print(f" {ticker} ({name}): {str(e)[:50]}")
            failed.append(ticker)
    
    # Combiner toutes les séries en DataFrame
    if all_series:
        prices = pd.concat(all_series, axis=1)
        
        # Normaliser l'index : garder seulement la date (sans l'heure)
        prices.index = prices.index.normalize()  # Supprime l'heure
        prices.index = prices.index.date  # Convertit en date pure
        prices.index.name = "Date"
        
        # Grouper par date et prendre la dernière valeur si doublons
        prices = prices.groupby(prices.index).last()
    else:
        prices = pd.DataFrame()
    
    print(f"\n{'='*60}")
    print(f"Téléchargements réussis : {len(all_series)}/{len(companies)}")
    if failed:
        print(f"Échecs ({len(failed)}) : {failed}")
    
    return prices, failed

def create_metadata_df():
    """Crée un DataFrame avec les métadonnées des entreprises"""
    
    metadata = []
    for name, info in companies.items():
        metadata.append({
            "Company": name,
            "Ticker": info["ticker"],
            "Country": info["country"],
            "Sector": info["sector"],
            "Treatment": 1 if info["sector"] == "Green" else 0
        })
    
    return pd.DataFrame(metadata)

def main():
    # Télécharger les prix
    prices, failed = download_all_prices()
    
    # Créer les métadonnées
    metadata = create_metadata_df()
    
    # Afficher un résumé
    print("\n" + "="*60)
    print("RÉSUMÉ")
    print("="*60)
    print(f"Entreprises téléchargées : {len(prices.columns)}/{len(companies)}")
    print(f"Période : {START_DATE} à {END_DATE}")
    print(f"Nombre d'observations : {len(prices)}")
    
    print(f"\nEntreprises par secteur :")
    print(metadata["Sector"].value_counts())
    
    print(f"\nEntreprises par pays :")
    print(metadata["Country"].value_counts())
    
    # Vérifier les données manquantes
    if not prices.empty and len(prices) > 0:
        missing = prices.isnull().sum()
        missing_pct = (missing / len(prices) * 100).round(1)
        print(f" Tickers avec >50% données manquantes :")
        problematic = missing_pct[missing_pct > 50]
        if len(problematic) > 0:
            for ticker, pct in problematic.items():
                print(f"   - {ticker}: {pct}% manquant")
        else:
            print("   Aucun")
    
    # Sauvegarder les fichiers
    prices.to_csv("stock_prices.csv")
    metadata.to_csv("companies_metadata.csv", index=False)
    
    print(" Fichiers sauvegardés :")
    print("   - stock_prices.csv (prix journaliers)")
    print("   - companies_metadata.csv (métadonnées)")
    
    return prices, metadata, failed

# =============================================================================
# EXÉCUTION
# =============================================================================

if __name__ == "__main__":
    prices, metadata, failed = main()
```
